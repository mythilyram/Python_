```python
# generate new columns 
# to find the age of the movie
#.apply applies transformations on the column
#lambda x is a way of quickly writing Python functions, like for loop, row-wise execution
df["age"] = df['release_year'].apply(lambda x: 2023 - x)
df.head()
```

```python
# derive a column using 2 existing columns
#profit = revenue - budget
df['profit'] = df.apply(lambda x: x['revenue']-x['budget'],axis=1)
#axis = 0 or 'index': apply fn to each column.
#axis = 1 or 'columns': apply fn to each row.
df.head()
```
